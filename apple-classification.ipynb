```python
df_numeric = df2.select_dtypes(include=['float64', 'int64'])


Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)


IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df3 = df2[~((df_numeric < lower_bound) | (df_numeric > upper_bound)).any(axis=1)]

print(df3)

```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
#GRADIENT BOOSTING

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model=GradientBoostingClassifier(n_estimators = 100,max_depth=3,learning_rate=0.1)

```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

imputer = SimpleImputer(strategy='mean')
pipeline = make_pipeline(imputer, model)
pipeline.fit(X_train, y_train)

```

```python
y_pred=pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
accuracy= accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
#grid search cv

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.7, 0.8, 1.0]
}
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameters: ", grid_search.best_params_)
```

```python
best_model = grid_search.best_estimator_

y_pred=pipeline.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```
